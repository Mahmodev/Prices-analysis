```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import NaN
```

```python
df1 = pd.read_csv('Datasets/elec_may_17/1.csv')
df2 = pd.read_csv('Datasets/elec_may_17/2.csv')
df3 = pd.read_csv('Datasets/elec_may_17/3.csv')
df4 = pd.read_csv('Datasets/elec_may_17/4.csv')
df5 = pd.read_csv('Datasets/elec_may_17/5.csv')
df6 = pd.read_csv('Datasets/elec_may_17/6.csv')
df7 = pd.read_csv('Datasets/elec_may_17/7.csv')
df8 = pd.read_csv('Datasets/elec_may_17/8.csv')
df9 = pd.read_csv('Datasets/elec_may_17/9.csv')
df10 = pd.read_csv('Datasets/elec_may_17/10.csv')
df11 = pd.read_csv('Datasets/elec_may_17/11.csv')
df12 = pd.read_csv('Datasets/elec_may_17/12.csv')
```

```python
gdf = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12])
```

```python
gdf
```

```python
gdf['Stars'] = gdf['Stars'].str.replace(' out of 5 stars','')
gdf['Price'] = gdf['Price'].str.replace('EGP', '')
gdf['Price'] = gdf['Price'].str.replace(',', '')
gdf['Price'] = gdf['Price'].str.replace('\xa0', '')
gdf['OldPrice'] = gdf['OldPrice'].str.replace('EGP', '')
gdf['OldPrice'] = gdf['OldPrice'].str.replace(',', '')
gdf['OldPrice'] = gdf['OldPrice'].str.replace('\xa0', '')
```

```python
df = gdf.drop_duplicates(subset=['Name', 'Price', 'OldPrice','Stars','BestSeller'], keep='first', inplace=False)
```

```python
df['Price'].unique()
```

```python
df['OldPrice'].unique()
```

```python
df = df[df.Price != 'Price'] # Delete Row has strings of the columns name
```

```python
df['Price'] = df["Price"].astype('float')
```

```python
df['OldPrice'] = df["OldPrice"].astype('float')
```

```python
df['Stars'] = df["Stars"].astype('float')
```

```python
df.info()
```

```python
df['Discount'] = (1-(df['Price']/df['OldPrice']))*100
df
```

```python
sorted_df = df.sort_values(by = ['Discount', 'Price'], ascending = False)
sorted_df.head(100)
```
