```python
df1 = pd.read_csv('Datasets/Fashion_May_24/fashion.csv')
df2 = pd.read_csv('Datasets/Fashion_May_24/2.csv')
```

```python
df = pd.concat([df1, df2])
```

```python
df['Stars'] = df['Stars'].str.replace(' out of 5 stars','')
df['Price'] = df['Price'].str.replace('EGP', '')
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].str.replace('\xa0', '')
df['OldPrice'] = df['OldPrice'].str.replace('EGP', '')
df['OldPrice'] = df['OldPrice'].str.replace(',', '')
df['OldPrice'] = df['OldPrice'].str.replace('\xa0', '')
```
